```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd
import pylab as pl
import numpy as np
import statistics
```

```python
# SQL Server

x = "Curso", "Departamento", "Instructor", "Persona"
# Tiempo en minutos
y1 = [61.33,81.12,52.55,168.47]
y2 = [40.29,68.10,41.47,138.37]

y1_num = len(y1)
indice_barras = np.arange(y1_num)
ancho_barras =0.35
 
plt.bar(indice_barras, y1, width=ancho_barras, label='Equipo 1')
plt.bar(indice_barras + ancho_barras, y2, width=ancho_barras, label='Equipo 2')
plt.legend(loc='best')

plt.xticks(indice_barras + ancho_barras, (x))
 
pl.xlabel('Tablas')
pl.ylabel('Tiempo requerido (minutos)')
pl.title('Inserción de medio millón de registros en SQL Server')
 
plt.show()
```
